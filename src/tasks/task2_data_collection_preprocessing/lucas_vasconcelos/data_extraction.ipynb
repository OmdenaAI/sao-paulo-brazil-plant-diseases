## Organize files by the disease since we receive all labels in the same path

```python
# Identify all paths available
paths = ['valid', 'train', 'test']
```

```python
#put the name of dataset downloadede by API
#dataset_name_downloaded = 'Doencas-1'
#dataset_name_downloaded = '1000101010001010101010111-3'
#dataset_name_downloaded = 'Penyakit_Daun_Kopi_Computer_Vision_Project'
#dataset_name_downloaded = 'Penyakit-Daun-NEW-30'
#dataset_name_downloaded = 'cafe-detect-16'
dataset_name_downloaded = 'your_dataset_name_downloaded'
```

```python
rows = []
# Iterate through each path
for path in paths:
    json_file = f'/content/{dataset_name_downloaded}/{path}/_annotations.coco.json'  # Update this with the correct path to your JSON file

    # Read the JSON file
    with open(json_file, 'r') as file:
        data = json.load(file)

    for i in range(len(data['images'])):
        image_id = data['images'][i]['id']
        category_id = data['annotations'][i]['category_id']
        file_name = data['images'][i]['file_name']
        height = data['images'][i]['height']
        width = data['images'][i]['width']

        rows.append([image_id, category_id, path, file_name, height, width])

# Create a DataFrame
df = pd.DataFrame(rows, columns=['id', 'category_id', 'path', 'file_name', 'height', 'width'])
```

```python
df_category = pd.DataFrame(data['categories'])
```

```python
df_category
```

```python
df_merged = pd.merge(df,df_category, left_on='category_id', right_on='id', how='left')
```

```python
df_merged
```

```python
df_merged['entire_path'] = dataset_name_downloaded + '/'  + df_merged['path'] + '/' + df_merged['file_name']
```
